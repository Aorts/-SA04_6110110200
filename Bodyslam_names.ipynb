```python
import lxml.etree
import requests

import pythainlp.tokenize
```

```python
resp = requests.get('https://เพลง.meemodel.com/ศิลปิน/บอดี้สแลม')

tree = lxml.etree.fromstring(resp.text, parser=lxml.etree.HTMLParser())
links = [f'https://เพลง.meemodel.com{path}' for path in tree.xpath('//table/tr/td[1]/a/@href')]
```

```python
names = set()

for link in links:
    resp = requests.get(link)

    tree = lxml.etree.fromstring(resp.text, parser=lxml.etree.HTMLParser())
    lyrics = ' '.join([line.strip() for line in tree.xpath('//*[@id="lyric"]//text()')])

    tokens = pythainlp.tokenize.word_tokenize(lyrics)

    names.update([t.strip() for t in tokens if len(t.strip()) > 1])
```

```python
len(names)
```

```python
step = 25
names = list(names)
for i in range(0, len(names), step):
    print(*names[i: i+step])
```
